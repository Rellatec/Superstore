```python
import pandas as pd

csv_file_path = 'data/Sample - Superstore 2.csv'

#  loading the CSV file with an alternative encoding
try:
    df = pd.read_csv(csv_file_path, encoding='latin1')
except UnicodeDecodeError:
    df = pd.read_csv(csv_file_path, encoding='ISO-8859-1')

```

```python
import matplotlib.pyplot as plt
import pandas as pd

df['Order Date'] = pd.to_datetime(df['Order Date'])

# Calculating monthly sales
monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()

# Plotting monthly sales
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

```python

import matplotlib.pyplot as plt
import pandas as pd

# Group data by 'Category' and sum up the profits
profit_by_category = df.groupby('Category')['Profit'].sum()

# Plotting profit by category
plt.figure(figsize=(10, 6))
profit_by_category.plot(kind='bar', color=['#00B7F1', 'lightgreen', '#353D45'])
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit ($)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate sales by customer segment
sales_by_segment = df.groupby('Segment')['Sales'].sum()

# Plotting sales by customer segment
plt.figure(figsize=(8, 8))
sales_by_segment.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#00B7F1', 'lightgreen', '#353D45'])
plt.title('Sales by Customer Segment')
plt.ylabel('')  # Removing y-label for clarity
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculating sales and profits by state and sorting
sales_by_state = df.groupby('State')['Sales'].sum().sort_values(ascending=False)
profits_by_state = df.groupby('State')['Profit'].sum().sort_values(ascending=False)

# Extracting top 5 states
top_5_states_by_sales = sales_by_state.head(5)
top_5_states_by_profit = profits_by_state.head(5)

# Plotting
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Top 5 States by Sales
top_5_states_by_sales.plot(kind='bar', ax=axes[0], color='#00B7F1')
axes[0].set_title('Top 5 States by Sales')
axes[0].set_xlabel('State')
axes[0].set_ylabel('Sales ($)')

# Top 5 States by Profit
top_5_states_by_profit.plot(kind='bar', ax=axes[1], color='lightgreen')
axes[1].set_title('Top 5 States by Profit')
axes[1].set_xlabel('State')
axes[1].set_ylabel('Profit ($)')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
orders_by_shipping_mode = df['Ship Mode'].value_counts()

# Plotting Orders by Shipping Mode
plt.figure(figsize=(10, 6))
orders_by_shipping_mode.plot(kind='bar', color=['#00B7F1', 'lightgreen', '#353D45', 'gold'])
plt.title('Orders by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```
